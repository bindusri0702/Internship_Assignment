## Individual LeaderBoard

```python
import re
# 'name' column is modified to give priority to alphabets rather than special characters
def order_string(text):
  text = text.upper()
  text = text[0] + re.sub(r'[^a-zA-Z]', '', text[1:])
  return(text)
```

### The individual leader board is ordered using the count of total statements and resaons given by a person in descending order, if this total count matches then they use alphabetical order prioritizing alphabets over special characters. 

```python
# it is always a good option to copy existing data frame rather than modifying it which can leads to some errors
r1 = rigor.copy()
# count of total statements and reasons given by a person
r1['total'] = r1['total_statements'] + r1['total_reasons']
# modify name prioritizing alphabets over special characters
r1['modified_name'] = r1['name'].apply(order_string)
```

```python
r2 = r1.sort_values(by=['total','modified_name'],ascending = [False,True], key=lambda x: [(not isinstance(val, str), val) for val in x])
```

```python
# reset index and drop unwanted columns
r2.reset_index(drop = True, inplace = True)
r2 = r2.drop(['S No','total','modified_name'], axis=1)
r2 = r2.reset_index(drop=True).rename(columns={'name' : 'Name','uid' : 'UID','total_statements' : 'No. of Statements','total_reasons' : 'No. of Reasons'})
# set rank column based on order
r2.set_index([pd.Index(list(range(1,len(r2)+1)))],inplace=True)	
r2 = r2.reset_index(drop=False).rename(columns={'index': 'Rank'})	
r2																		
```

## Check if given output matches with computed output

```python
# Check if the two DataFrames are equal
if r2.equals(Individual_leader_board):
    print("The two DataFrames are equal")
else:
    print("The two DataFrames are not equal")
```

## TeamWise LeaderBoard

```python
# copy user_ids input
u1 = User_Ids.copy()
# brandtech lab is written in two forms even though both indicating same team, so modify that to a single name
u1['Team Name'] = u1['Team Name'].replace('Brandtech Lab','BrandTech Lab')
# merge 2 dataframes based on 'name' column
user_team = pd.merge(u1, rigor, left_on='Name', right_on='name')
# group by team name
team = user_team.groupby('Team Name',as_index = False).mean()
# rename columns as given in output
team = team.rename(columns={'Team Name':'Thinking Teams Leaderboard','total_statements' : 'Average Statements', 'total_reasons' : 'Average Reasons'})
# drop unwanted columns
team = team.drop(['S No_x','User ID','S No_y','uid'], axis=1)
# sort by average statements made by each group and then by average reasons in descending order
team = team.sort_values(by=['Average Statements','Average Reasons'],ascending = [False,False])
# reset the index as team rank
team.set_index([pd.Index(list(range(1,len(team)+1)))],inplace=True)	
team = team.reset_index(drop=False).rename(columns={'index': 'Team Rank'})
team
```

```python
# Check if the two DataFrames are equal
if team.equals(team_leader_board):
    print("The two DataFrames are equal")
else:
    print("The two DataFrames are not equal")
```

### The above result shows the two dataframes are not equal this is because the average reasons of 'Tech Lab' doesn't match in both dataframes.
### According to the computed dataframe (team) the average reasons is the avergae of total reasons given by each person from the same team. 
### In this case imshawan, Ronak 0, Shagun belongs to Tech Lab team and they made 9, 19, 2 reasons respectively.
### Average_reasons = (9+19+2)/3 = 30/3 = 10
### But the given output contains 9.33 instead of 10, so this is the reason for difference in datframes.
